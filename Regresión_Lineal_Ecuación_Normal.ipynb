```
#Importando librerias neceserias
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

```
# Usando un data set de sklearn
from sklearn.datasets import load_boston
```

```
# Cargando la libreria
boston = load_boston()

print(boston.DESCR)
```

```
#Delaracion de matrices

#Numero medio de habitaciones por cada barrio
X = np.array(boston.data[:, 5])

#Valor medio de habitaciones por cada barrio
Y = np.array(boston.target)
```

Formula para minimizar el error cuadratico medio (MCO): 
$\beta = (X^TX)^{-1}X^TY$

```
#Graficar datos
plt.scatter(X, Y, alpha=0.3)
```

```
# shape de X
X_shape = np.array(X.shape)
print(X_shape)
```

```
# Añadimos columna de 1s para termino independiente
X_1 = np.array([np.ones(X_shape[0]), X]).T
print(X_1)
```

```
# Ecuacion Normal
B = np.linalg.inv(X_1.T @ X_1) @ X_1.T @ Y

# Graficar Linea
def linearEquations(x):
  return B[1] * x + B[0]

x = np.array(range(4,10))
plt.scatter(X, Y, alpha=0.3)
plt.plot(x,[linearEquations(i) for i in x], c="red")
plt.show()
```
